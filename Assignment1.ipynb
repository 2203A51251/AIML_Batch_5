```python
import pandas as pd

people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

healthcare_data = pd.DataFrame({
    'Name': people,
    'Age': age,
    'Height': height,
    'Weight': weight
})

print(healthcare_data)
```

2. Using the generated dataset given above. Generate the following graphs and justify the
relationships among the vectors

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

3.Scatter plot

```python
import matplotlib.pyplot as plt

your_x_data = [1, 2, 3, 4, 5]
your_y_data = [2, 4, 1, 3, 5]

plt.scatter(your_x_data, your_y_data, label='Scatter Plot')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

4.Bar Graph

```python
import matplotlib.pyplot as plt

your_categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
your_values = [10, 24, 15, 30, 18]

plt.bar(your_categories, your_values, color='blue', alpha=0.7)
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

5.Histogram

```python
import matplotlib.pyplot as plt

your_data = [2, 3, 3, 4, 4, 4, 5, 5, 6, 7, 7, 7, 7, 8, 8, 8, 8, 9, 9, 10]

plt.hist(your_data, bins=10, color='gray', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```
